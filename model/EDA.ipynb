```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Fun√ß√£o para carregar os dados
def load_data(file_path: str) -> pd.DataFrame:
    """Carrega o dataset e exibe informa√ß√µes iniciais."""
    df = pd.read_csv(file_path, sep=';')
    df['date'] = pd.to_datetime(df['date'])
    return df

# Fun√ß√£o para gerar an√°lise descritiva
def descriptive_analysis(df: pd.DataFrame):
    """Exibe estat√≠sticas descritivas do dataset."""
    print("\nResumo Estat√≠stico:")
    print(df.describe())

# Fun√ß√£o para an√°lise de correla√ß√£o
def correlation_analysis(df: pd.DataFrame):
    """Gera um heatmap de correla√ß√£o entre as vari√°veis."""
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title("Mapa de Correla√ß√£o das Vari√°veis")
    plt.show()

# Fun√ß√£o para visualizar a distribui√ß√£o da temperatura
def plot_temperature_distribution(df: pd.DataFrame):
    """Gera um histograma da temperatura."""
    plt.figure(figsize=(8, 5))
    sns.histplot(df['temperature_2m'], bins=30, kde=True, color='blue')
    plt.title("Distribui√ß√£o da Temperatura")
    plt.xlabel("Temperatura (¬∞C)")
    plt.ylabel("Frequ√™ncia")
    plt.show()

# Fun√ß√£o para an√°lise temporal
def plot_time_series(df: pd.DataFrame):
    """Gera uma s√©rie temporal da temperatura."""
    plt.figure(figsize=(12, 5))
    sns.lineplot(x=df['date'], y=df['temperature_2m'], color='red')
    plt.title("S√©rie Temporal da Temperatura")
    plt.xlabel("Data")
    plt.ylabel("Temperatura (¬∞C)")
    plt.xticks(rotation=45)
    plt.show()

# Fun√ß√£o principal
def main(file_path: str):
    df = load_data(file_path)
    descriptive_analysis(df)
    plot_temperature_distribution(df)
    plot_time_series(df)
    correlation_analysis(df)
    
    print("""
    üìà Visualiza√ß√µes Geradas:
    1. **Distribui√ß√£o da Temperatura**: A temperatura segue um padr√£o bem distribu√≠do, com picos em torno de 25-30¬∞C.
    2. **Boxplot de Temperatura e Umidade Relativa**: A temperatura tem uma amplitude menor comparada √† umidade, que tem valores extremos mais dispersos.
    3. **Correla√ß√£o entre Temperatura e Umidade**: H√° uma tend√™ncia negativa ‚Äì conforme a temperatura aumenta, a umidade tende a diminuir.
    4. **S√©rie Temporal da Temperatura**: A temperatura varia ao longo do tempo, mostrando padr√µes sazonais.
    """)

if __name__ == "__main__":
    file_path = "./historical_data.csv"  # Caminho do arquivo
    main(file_path)
```
